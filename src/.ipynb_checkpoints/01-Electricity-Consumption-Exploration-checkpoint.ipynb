In this exercise, you will be asked to explore time series data in several ways.

You will be working with time series of Open Power System Data ([OPSD](https://open-power-system-data.org/)) for Germany. 

The data set includes daily electricity consumption, wind power production, and solar power production between 2006 and 2017. 
```
- Date — The date (yyyy-mm-dd format)
- Consumption — Electricity consumption in GWh
- Wind — Wind power production in GWh
- Solar — Solar power production in GWh
- Wind+Solar — Sum of wind and solar power production in GWh
```

# Data Exploration

````python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from IPython.display import Image
````

````python
# Load file
df = pd.read_csv("opsd_germany_daily.csv", parse_dates=['Date'], index_col='Date')
````

````python
# Description
df.head()
````

````python
df.info()
````

Then, make sure to transform the dates into datetime format in pandas :

````python
# Transform the dates into datetime format
df.index = pd.to_datetime(df.index)
````

````python
df.index
````

### 1. Plotting and understanding a time series

Q1 - Plot the different time series. What do you understand, what do you notice ?

````python
plt.figure(figsize=(14,10))
plt.plot(df["Consumption"], label="Consumption", linewidth=0.5)
plt.plot(df["Wind+Solar"], label="Wind+Solar", linewidth=0.5)

plt.title("Electricity consumption and production comparison in Germany between 2012 and 2017")
plt.legend()
plt.show()
````

We can also avoid plotting lines and simply plot dots for the consumption :

````python
plt.figure(figsize=(14,10))
plt.plot(df['Consumption'], label="Consumption", linewidth=0.5)
plt.title("Daily electricity consumption in Germany between 2006 and 2017")
plt.legend()
plt.show()
````

And plot the production over time :

````python
plt.figure(figsize=(14,10))
plt.plot(df['Wind+Solar'], label="Wind+Solar")

plt.title("Daily wind and solar power production in Germany between 2006 and 2017")
plt.legend()
plt.show()
````

What can we notice ?

How do consumption and production compare ?

The consumption is much higher than the production. This implies that the main source of electricity consumption for German does not come from wind or sun. However, we can notice a correlative increase in these two types of energy over time. It can be assumed that this trend will continue to grow in coming years.

In other hand, we can see a correlation between consumption and production : more consumption and production in winter time, and less in summer.

### 2. Change the scale

Q2 - Try to change the scale to understand the recurrent patterns !

````python
# Plot over the year

plt.figure(figsize=(12,8))
plt.plot(df["2012"]["Consumption"])
plt.plot(df["2012"]["Wind+Solar"])

plt.title("Electricity consumption and production comparison in 2012 in Germany")
plt.legend()
plt.show()
````

It confirms the trend : more consumption in winter time, and less in summer. We can then zoom on a given month :

````python
# Plot over a month

plt.figure(figsize=(12,8))
plt.plot(df["2012-05"]["Consumption"], label="Consumption")
plt.plot(df["2012-05"]["Wind+Solar"], label="Wind+Solar")

plt.title("Electricity consumption and production comparison in May, 2012 in Germany")
plt.legend()
plt.show()
````

When zooming on a month, we oberve another seasonality. There seems to be days of the week in which people consume less. For example, here's what happens when we highlight Sundays :

````python
# Plot on Sundays

plt.figure(figsize=(14,10))
plt.plot(df["2013-03"]["Consumption"], label="Consumption")
plt.plot(df["2013-03"][df["2013-03"].index.dayofweek==6]["Consumption"], label="Sundays", linestyle="None", marker="o", color="red")

plt.title("Electricity consumption on Sundays in march 2013 in Germany")
plt.legend()
plt.show()
````

### 3. Further exploration

Q3 - To further explore the dataset, we might need to group the data. Create new columns for the consumption (by year and month), and create box-plots with seaborn. Comment on your results.

````python
# new columns 'year' and 'month'

df["Year"] = df.index.year
df["Month"] = df.index.month

df.head()
````

````python
# Boxplot year
import seaborn as sns

plt.figure(figsize=(20,10))
sns.boxplot(x=df["Year"], y=df["Consumption"], data=df)
plt.title("Consumption in Germany by year between 2006 and 2017")
````

````python
# Boxplot month
plt.figure(figsize=(20,10))
sns.boxplot(x=df["Month"], y=df["Consumption"], data=df)
plt.title("Consumption in Germany by month between 2006 and 2017")
````
